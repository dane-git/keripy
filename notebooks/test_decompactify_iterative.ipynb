```python
from copy import deepcopy

## hacky Convenience functions to get the partial version string --> length

B64_VALUES = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789-_"

def value_of(ch):
  return B64_VALUES.index(ch)

def b64_to_int(value):
  result = 0

  # Iterate over each character in the Base64 string
  for ch in value:
      result <<= 6  # Shift left by 6 bits
      result |= value_of(ch)  # OR with the Base64 value of the character

  return result

def int_to_b64(value, length=4):
    """
    Converts an integer to a Base64 string of a specified length using the custom Base64 set.
    Pads the front with 'A' if necessary to achieve the desired length.

    Parameters:
        value (int): The integer to convert.
        length (int): The desired length of the Base64 string.

    Returns:
        str: The Base64 string representation of the integer, padded to the specified length.
    """
    if value < 0 or value >= (1 << (6 * length)):  # Ensure the integer fits in the specified length
        raise ValueError(f"Integer value out of range for {length} Base64 characters")

    # Convert integer to Base64 characters
    b64_string = ""
    for _ in range(length):
        b64_string = B64_VALUES[value & 0x3F] + b64_string  # Get last 6 bits and prepend character
        value >>= 6  # Shift value right by 6 bits for the next character

    # If the resulting string is shorter than the desired length, pad with 'A' at the front
    return b64_string.rjust(length, "A")

def update_v2string_length(ked):
    _raw = coring.dumps(ked)
    _size = len(_raw)
    size_b64 = int_to_b64(_size)
    ## SANITY CHECK
    size_from_b64 = b64_to_int(size_b64)
    # print(f'byte size: {_size} \t\tsize_b64: {size_b64}, {size_from_b64 == _size}')
    assert (size_from_b64 == _size)
    ## update v string:
    v_string = ked['v'][:-5] + size_b64 + '.'
    return v_string
    
# ==================================== SIMPLE DECOMPACTIFY ( iterative ) =================================================

## construct a partial, given a list of paths and all the compact sads.
def contruct_partial(paths, sads):
    # GET ROOT KED ( COMAPCITIFED ) 
    for sad in sads:
        if sad.path == [sad.label]:
            root_ked = deepcopy(sad.ked)

    # sort from shortest to longest, just because
    paths = sorted(paths, key=len)
    for path in paths:
        root_ked = simple_decompactify(path, sads, root_ked)
    return root_ked
    
def vIsFirst(data, key='v'):
    if not isinstance(data, dict):
        raise ValueError("Input data must be a dictionary.")
    
    keys = list(data.keys())
    return keys[0] == key if keys else False
    
def simple_decompactify(path, all_sads, root_ked = {}):
    """
    Rebuilds a nested structure recursively by adding compactified versions
    of SADs along the specified path.

    Parameters:
        path (list): The path to decompactify to.
        all_sads (list): A list of Sadder instances.
        root_ked (dict): the root ked to build on top of. default will fill to the compact ked
    Returns:
        dict: The modified root object with the specified path decompactified.
    """    
    print("PATH", path)
    if root_ked == {}:
        for sad in all_sads:
            if [sad.label] == sad.path:
                rk = deepcopy(sad.ked)
                root_ked = deepcopy(rk)
    match_path = []
    for key in path:
        # go to current spot
        match_path.append(key)
        if key == all_sads[0].label:
            if vIsFirst(root_ked):
                root_ked['v']  = update_v2string_length(root_ked)
            return root_ked
        current_obj = root_ked
        for p in match_path:
            if p in current_obj:
                if isinstance(current_obj[p], (dict, list)):
                    current_obj  = current_obj[p]
                    continue
                else:
                    ## find sad that matches:
                    for sad in all_sads:
                        if sad.path[:-1] ==  match_path:
                           current_obj[p] = deepcopy(sad.ked)
                    continue
            else:
                print(p, match_path)
                for sad in all_sads:
                    if sad.path[:-1] ==  match_path:
                        current_obj[p] = deepcopy(sad.ked)
                        break
    if vIsFirst(root_ked):
        root_ked['v']  = update_v2string_length(root_ked)
    return root_ked                            
                
# ============================================================================================================
```

```python
def compare_structure(obj1, obj2, path=""):
    """
    Compare the structure and keys of two objects and report differences.

    Parameters:
        obj1: The first object to compare.
        obj2: The second object to compare.
        path: The current path being inspected (used for recursive calls).

    Returns:
        list: A list of differences, where each difference is a string describing the discrepancy.
    """
    differences = []

    if isinstance(obj1, dict) and isinstance(obj2, dict):
        # Compare dictionary keys
        keys1 = set(obj1.keys())
        keys2 = set(obj2.keys())
        
        # Find missing keys in each dictionary
        missing_in_obj2 = keys1 - keys2
        missing_in_obj1 = keys2 - keys1

        for key in missing_in_obj2:
            differences.append(f"Key '{key}' missing in second object at path '{path}'")
        for key in missing_in_obj1:
            differences.append(f"Key '{key}' missing in first object at path '{path}'")

        # Recursively compare keys present in both
        for key in keys1 & keys2:
            new_path = f"{path}.{key}" if path else key
            differences.extend(compare_structure(obj1[key], obj2[key], new_path))

    elif isinstance(obj1, list) and isinstance(obj2, list):
        # Compare list lengths
        if len(obj1) != len(obj2):
            differences.append(f"List length mismatch at path '{path}' (obj1: {len(obj1)}, obj2: {len(obj2)})")

        # Compare list items pairwise
        for i, (item1, item2) in enumerate(zip(obj1, obj2)):
            new_path = f"{path}[{i}]"
            differences.extend(compare_structure(item1, item2, new_path))

    else:
        # Check if types match
        if type(obj1) != type(obj2):
            differences.append(f"Type mismatch at path '{path}' (obj1: {type(obj1)}, obj2: {type(obj2)})")

    return differences
```

```python
def get_differences(obj1, obj2, path=""):
    """
    Compare two nested objects (e.g., dictionaries, lists) and report differences.

    Parameters:
        obj1: The first object to compare.
        obj2: The second object to compare.
        path: The current path in the nested structure (used for recursion).

    Returns:
        A dictionary containing the differences, keyed by their paths.
    """
    differences = {}

    if type(obj1) != type(obj2):
        differences[path or "."] = {"type1": type(obj1).__name__, "type2": type(obj2).__name__}
        return differences

    if isinstance(obj1, dict):
        all_keys = set(obj1.keys()).union(set(obj2.keys()))
        for key in all_keys:
            new_path = f"{path}.{key}" if path else key
            if key not in obj1:
                differences[new_path] = {"missing_in_obj1": obj2[key]}
            elif key not in obj2:
                differences[new_path] = {"missing_in_obj2": obj1[key]}
            else:
                differences.update(get_differences(obj1[key], obj2[key], new_path))

    elif isinstance(obj1, list):
        len1, len2 = len(obj1), len(obj2)
        if len1 != len2:
            differences[path or "."] = {"length1": len1, "length2": len2}

        min_len = min(len1, len2)
        for index in range(min_len):
            new_path = f"{path}[{index}]"
            differences.update(get_differences(obj1[index], obj2[index], new_path))

        for index in range(min_len, len1):
            differences[f"{path}[{index}]"] = {"missing_in_obj2": obj1[index]}

        for index in range(min_len, len2):
            differences[f"{path}[{index}]"] = {"missing_in_obj1": obj2[index]}

    else:
        if obj1 != obj2:
            differences[path or "."] = {"value1": obj1, "value2": obj2}

    return differences
```
